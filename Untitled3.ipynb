```python
try:
    from num2words import num2words
except:
    !pip install num2words
    from num2words import num2words
try:
    from wordcloud import WordCloud
except:
    !pip install wordcloud
    from wordcloud import WordCloud
try:
    from faker import Faker
except:
    !pip install faker
    from faker import Faker
import random, ast
import matplotlib.pyplot as plt

#Gera o objeto fake usando a classe Faker da biblioteca faker
fake = Faker('pt_BR')

#Gera um dicionario com uma quantidade determinada de nomes aleatorios e pontos entre 0 e 20
quantidade = 50
notaspossiveis = 20
lista = {fake.name(): random.randint(0,notaspossiveis) for i in range(quantidade)}

#Salva o arquivo txt
open('nomes.txt', 'w').write(str(lista))   

#Lê o arquivo txt
lista = ast.literal_eval(open('nomes.txt', 'r').read())

#Gera o Histograma
lista = [lista[i] for i in lista]
maior = 0
for i in lista:
    vezes = lista.count(i)
    if vezes > maior:
        maior = vezes

plt.hist(lista, notaspossiveis, facecolor='blue', alpha=0.8, edgecolor='#000', linewidth=1.2)
plt.xlabel('Notas')
plt.ylabel('Quantidade')
plt.title('Notas dos alunos')

#Melhorias visuais do histograma
plt.gca().yaxis.grid(True, color='#000', linestyle='--', linewidth=1)
plt.xticks([i for i in range(notaspossiveis + 1)])
plt.xlim([0, notaspossiveis])
plt.yticks([i for i in range(maior + 2)])
plt.ylim([0, maior + 1])
plt.show()

#Gera a nuvem de palavras
lista = ','.join([num2words(i, lang='pt') for i in lista])
lista = lista.replace('deza', 'deze') #Corrige falhas na tradução do num2words
nuvem = WordCloud(background_color='#fff').generate(lista)

print('') #Espaçamento entre histograma e nuvem de palavras

#Melhorias visuais da nuvem de palavras
plt.figure()
plt.axis("off")
plt.imshow(nuvem, interpolation="bilinear") #Nuvem de palavras
plt.show()
pass #Apenas pasa não printar um objeto do matplotlib
```
